-The observed trend

The first major observable trend in urban data sets generated by uniform sampling within the full latitude and longitude of the global coordinate system is that cities seem to be disproportionately distributed in hemispheres-that is, the northern hemisphere usually contains 65% of the cities in the data set, while the remaining 30% are located in the southern hemisphere. At first glance, this distribution seems uneven until we consider that most of the earth's land is in the northern hemisphere, which includes about 71% of the world's land (see here). This directly corresponds to the percentage of urban distribution in the hemisphere, which confirms that the latitude distribution of sampled cities is relatively uniform. It is expected that cities will account for half of the surface area above and below the equator, but finding different results will make us question why there is such a difference. the answer seems to be the percentage of land area in each hemisphere. because cities are mainly built on the land. 

The second trend observed is that temperatures rise linearly at latitudes of 20 degrees in the northern and southern hemispheres rather than the peak at the equator. This can be explained by the inclination of the earth's axis, which is about 23 degrees (see here). When the temperature is compared with latitude, the r square values of the best fit regression lines in the northern and southern hemispheres are about-0.7 and 0.8, respectively, and these correlations may increase if the data are separated by latitude 23 rather than the equator (especially for the northern hemisphere because the graph includes some data before the peak temperature is reached, which confuses the linear best fit). These R values show that there is a strong linear relationship between latitude and temperature in the northern and southern hemispheres. The 23-degree latitude line represents the point closest to the sun on Earth during summer in the northern hemisphere, and we can expect that the-23 latitude line will show the highest temperature in the world during winter in the northern hemisphere and summer in the southern hemisphere. 

The third trend observed from the Vacation Py notebook is that humidity values seem to be high all over the world, and many of the red areas on the heat map represent the maximum or near-maximum humidity in the area shown. Compared with the inland areas deep in the continent, these values seem to be particularly high and sustained near coasts and some large waters, such as rivers or lakes. This may be because there is more evaporated water in the air near large areas of water, which can prevent the climate from becoming too dry. In addition to the relative humidity percentage, it would be interesting to use statistics that measure absolute humidity, such as measuring water saturation in the air, which is based on water mass per unit volume or other absolute characteristics. Relative humidity varies with temperature, so warmer air can hold more moisture than cooler air, which will make absolute measurements of humidity show higher values in hot and humid areas such as the tropics or rainforests. rather than dryers or temperate zones where the relative humidity is still relatively high.
```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import scipy.stats as st
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# # Output File (CSV)
output_data_file = "output_data/cities.csv"

# # Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Create empty lists to store the data for each field
city_list = []
cloud_list = []
country_list = []
date_list = []
humidity_list = []
lats_list = []
lngs_list = []
temp_max_list = []
wind_speed_list = []
index_counter = 0
set_counter = 1
print("Beginning Data Retrieval ")
print("-------------------------------")
# Set base url and units variables, and combine them to get the query url
base_url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
query_url = f"{base_url}appid={weather_api_key}&units={units}&q="
# Loop through cities
#For loop matching city names with city_list using enumerate method
for index, city in enumerate(cities, start = 1):
    try:
        response = requests.get(query_url + city).json()
        city_list.append(response["name"])
        cloud_list.append(response["clouds"]["all"])
        country_list.append(response["sys"]["country"])
        date_list.append(response["dt"])
        humidity_list.append(response["main"]["humidity"])
        lats_list.append(response["coord"]["lat"])
        lngs_list.append(response["coord"]["lon"])
        temp_max_list.append(response['main']['temp_max'])
        wind_speed_list.append(response["wind"]["speed"])
        
        if index_counter > 49:
            index_counter = 0
            set_counter = set_counter + 1
        
        else:
            index_counter = index_counter + 1
        
        print(f"Processing Record {index_counter} of Set {set_counter} : {city}")
        
    except(KeyError, IndexError):
        print("City not found. Skipping...")
        
print("-------------------------------")
print("Data Retrieval Complete")
print("-------------------------------")
        
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#Create a dataframe using information from data retrieval
weather_data = pd.DataFrame({
    "City" : city_list,
    "Lat" : lats_list,
    "Lng" : lngs_list,
    "Max Temp" : temp_max_list,
    "Humidity" : humidity_list,
    "Clouds" : cloud_list,
    "Wind Speed" : wind_speed_list,
    "Country" : country_list,
    "Date" : date_list    
})
#Count weather data
weather_data.count()
#Display dataframe
weather_data
```

## Linear Regression

```python
# Create Northern and Southern Hemisphere DataFrames
northern_hemisphere = weather_data.loc[weather_data["Lat"] >= 0]
southern_hemisphere = weather_data.loc[weather_data["Lat"] < 0]
# Define function for creating linear agression and scatter plot
def linear_agression(x,y):
    print(f"The r-squared is : {round(st.pearsonr(x, y)[0],18)}")
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
    regress_values = x * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    plt.scatter(x, y)
    plt.plot(x,regress_values,"r-")
    return line_eq
# Define a fuction for annotating
def annotate(line_eq, a, b):
    plt.annotate(line_eq,(a,b),fontsize=15,color="red")
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
equation = linear_agression(northern_hemisphere["Lat"], northern_hemisphere["Max Temp"])
annotate(equation, 0, 0)
plt.xlabel("Latitude")
plt.ylabel("Max Temp ")
plt.savefig("../Images/Northern Hemisphere - Max Temp vs. Latitude Linear Regression.png")
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
equation = linear_agression(southern_hemisphere["Lat"],southern_hemisphere["Max Temp"])
annotate(equation, -30, 50)
plt.xlabel("Latitude")
plt.ylabel("Max Temp (F)")
plt.savefig("../Images/Southern Hemisphere - Max Temp vs. Latitude Linear Regression.png")
```
